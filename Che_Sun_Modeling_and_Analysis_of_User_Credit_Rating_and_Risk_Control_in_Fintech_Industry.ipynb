## Part 2: Data cleaning and Feature Preprocessing

## **Data Cleaning**

```python
# Data Cleaning / Integration
class_mapping = {label:idx for idx, label in enumerate(np.unique(CheSun_LC_df['term']))}
CheSun_LC_df['term']=CheSun_LC_df['term'].map(class_mapping)
class_mapping = {label:idx for idx, label in enumerate(np.unique(CheSun_LC_df['grade']))}
CheSun_LC_df['grade']=CheSun_LC_df['grade'].map(class_mapping)
class_mapping = {label:idx for idx, label in enumerate(np.unique(CheSun_LC_df['home_ownership']))}
CheSun_LC_df['home_ownership']=CheSun_LC_df['home_ownership'].map(class_mapping)

class_mapping = {label:idx for idx, label in enumerate(np.unique(CheSun_LC_df['verification_status']))}
CheSun_LC_df['verification_status']=CheSun_LC_df['verification_status'].map(class_mapping)

class_mapping = {label:idx for idx, label in enumerate(np.unique(CheSun_LC_df['purpose']))}
CheSun_LC_df['purpose']=CheSun_LC_df['purpose'].map(class_mapping)

class_mapping = {label:idx for idx, label in enumerate(np.unique(CheSun_LC_df['addr_state']))}
CheSun_LC_df['addr_state']=CheSun_LC_df['addr_state'].map(class_mapping)

class_mapping = {'Fully Paid' : 0, 'Charged Off' : 1}
CheSun_LC_df['loan_status']=CheSun_LC_df['loan_status'].map(class_mapping)

CheSun_LC_df.head()
```

```python
# deal with missing value
CheSun_LC_df = CheSun_LC_df.select_dtypes(include=[np.number]).interpolate().dropna()
# drop high correlation and high variance colums
CheSun_LC_df = CheSun_LC_df.drop(["total_pymnt"], axis=1)
CheSun_LC_df = CheSun_LC_df.drop(["total_pymnt_inv"], axis=1)
CheSun_LC_df = CheSun_LC_df.drop(["total_rec_int"], axis=1)
CheSun_LC_df = CheSun_LC_df.drop(["id"], axis=1)
CheSun_LC_df = CheSun_LC_df.drop(["total_rec_prncp"], axis=1)

CheSun_LC_df.head(10)
```

## **Data Split**

```python
# split dataset
from sklearn.model_selection import train_test_split

yPredict = CheSun_LC_df.loan_status
XClean = CheSun_LC_df.drop(["loan_status"], axis=1)

# Reserve 25% for testing
X_train, X_test, y_train, y_test = train_test_split(XClean, yPredict, random_state=42, test_size=.25)
```

```python
X_train.head()
```

```python
y_train.head()
```

## **Data preprocessing**

```python
# transform file
# standardization
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

## Part 3: Model Training and Selection

## **3.1: Logistic/ K Nearest Neighbors/ LogisticRegression**

### 3.1.1 Build Models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier

# Logistic Regression
classifier_logistic = LogisticRegression()

# K Nearest Neighbors
classifier_KNN = KNeighborsClassifier()

# Random Forest
classifier_RF = RandomForestClassifier()
```

```python
classifier_logistic.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
prediction_Regression = classifier_logistic.predict(X_test)
print(classification_report(y_test,prediction_Regression))
```

```python
classifier_KNN.fit(X_train, y_train)
```

```python
classifier_RF.fit(X_train, y_train)
```

```python
prediction_RF = classifier_RF.predict(X_test)
print(classification_report(y_test,prediction_RF))
```

```python
print(prediction_RF)
```

## **3.1.2 Cross Validation**

```python
# Use 5-fold Cross Validation to get the accuracy for different models
from sklearn import model_selection

model_names = ['Logistic Regression','KNN','Random Forest']
model_list = [classifier_logistic, classifier_KNN, classifier_RF]
count = 0

for classifier in model_list:
    cv_score = model_selection.cross_val_score(classifier, X_train, y_train, cv=5)
    print(cv_score)
    print('Model accuracy of ' + model_names[count] + ' is ' + str(cv_score.mean()))
    count += 1
```

**Logistic Regression has the best performance in these three **

## **3.2 Other model - SVM**

```python
from sklearn.svm import SVC

classifier_SVC = SVC()

cv_score = model_selection.cross_val_score(classifier_SVC, X_train, y_train, cv=5)
print('Model accuracy of SVM is: ' + str(cv_score.mean()))
```
